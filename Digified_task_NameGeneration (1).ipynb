```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import os 
import random
```

```python
data=pd.read_csv("/content/Arabic_names.csv")
```

```python
data.head()
```

```python
Male_ds=data[data.Gender=='M']
Female_ds=data[data.Gender=='F']
```

```python
def Generate_Data_Names(no_of_name,Gender,Fake=False):
      names=[]
      uglyfication_list=['#','@',str(np.random.choice(10)),'*','&']
      for i in range(no_of_name):
        if Gender=='M':
          ds=Male_ds
        else:
          ds=Female_ds

        idx_first=np.random.choice(len(ds))
        idx_mid=np.random.choice(len(Male_ds))
        idx_last=np.random.choice(len(Male_ds))
        first=ds.iloc[idx_first][0]
        mid=Male_ds.iloc[idx_mid][0]
        last=Male_ds.iloc[idx_last][0]
        label=np.ones(no_of_name).astype(np.int8)

        if Fake:
          first=''.join(random.sample(first,len(first)))
          mid=''.join(random.sample(mid,len(mid)))
          last=''.join(random.sample(last,len(last)))

          if np.random.uniform()>0.8:
            first=first+np.random.choice(uglyfication_list)
          if np.random.uniform()>0.8:
            last=last+np.random.choice(uglyfication_list)
          if np.random.uniform()>0.8:
            mid=mid+np.random.choice(uglyfication_list)
          label=np.zeros(no_of_name).astype(np.int8)

          

        full_name= first+' '+mid+' '+last
        names.append(full_name)
      
      d=np.array([names,label])
      data=pd.DataFrame(d.T)
      data.columns=['names','labels']
      return data
```

```python
correct_names_male=Generate_Data_Names(50000,'M')
correct_names_female=Generate_Data_Names(50000,'F')
Correct_ds=pd.concat([correct_names_male,correct_names_female],axis=0)
```

```python
Correct_ds.sample(10)
```

```python
fake_names_male=Generate_Data_Names(50000,'M',Fake=True)
fake_names_female=Generate_Data_Names(50000,'F',Fake=True)
Fake_ds=pd.concat([fake_names_male,fake_names_female],axis=0)
```

```python
Fake_ds.sample(10)
```

```python
All_data=pd.concat([Correct_ds,Fake_ds],axis=0)
All_data.to_csv("Names_Verification_ds_.csv")
```
